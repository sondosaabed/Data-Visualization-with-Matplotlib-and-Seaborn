# L6 Faceting in Two Directions

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

Faceting can be extended to multivariate cases by having each facet be a bivariate plot instead.

Let's start by doing our normal imports of the fuel economy dataset, and take our normal sample to simplify our visualizations.

```python
# load dataset
df = pd.read_csv('../../Data/fuel-econ.csv')

# create transmission type column
df['trans_type'] = df['trans'].apply(lambda x:x.split()[0])
```

From the code we started off with, we can just change the mapping parameters to a scatter plot instead of histogram, and provide two axes variables instead of just one.

```python
g = sns.FacetGrid(data=df, col='VClass')
g.map(plt.scatter, 'displ','comb')
```

We're not limited to only facet by one variable either: we can add a `row` parameter to facet by row.

```python
# Scatterplot Fuel Efficiency by Vehicle Class and Transmission Type
g = sns.FacetGrid(data=df, col='VClass', row='trans_type')
g.map(plt.scatter, 'displ','comb')
```

Use the `margin_titles=True` to move the row titles to the rows for easier reading.

```python
# Scatterplot Fuel Efficiency by Vehicle Class and Transmission Type
g = sns.FacetGrid(data=df, col='VClass', row='trans_type', margin_titles=True)
g.map(plt.scatter, 'displ','comb')
```
